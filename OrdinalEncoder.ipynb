```python
import numpy as np
import pandas as pd 
from sklearn.preprocessing import OrdinalEncoder
```

```python
data = pd.read_csv("C:/Users/patan/DS/Machine Learning/Dataset/customer.csv")
```

```python
data
```

```python
new_data = data[["gender","education","purchased"]]
new_data.head()
```

```python
oe = OrdinalEncoder()
```

```python
feature_array = oe.fit_transform(new_data)
feature_array
```

```python
data[["gender","education","purchased"]] = feature_array
```
