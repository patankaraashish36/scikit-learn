#  Manually OE

```python
cars["fuel"].unique()
```

```python
fuel = ["Petrol", "Diesel" ,"CNG","LPG"]
```

```python
# cars["owner"].unique()
```

```python
# owner = ["Test Drive Car", "First Owner","Second Owner" ,"Third Owner"]
```

```python
moe = OrdinalEncoder(categories=[fuel])
```

```python
ecd = moe.fit_transform(cars[["fuel"]])
ecd
```

```python
cars["fuel"] = ecd
```
