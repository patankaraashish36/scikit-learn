#  Mean Imputaion

```python
mean_age = X_train["Age"].mean()
mean_age
```

```python
X_train["Mean_Age"]  = X_train["Age"].fillna(mean_age)
```

```python
X_train
```

#  Check changes in Distribution

```python
fig = plt.figure()
ax = fig.add_subplot(111)
X_train["Age"].plot(kind='kde', ax=ax)
X_train["Mean_Age"].plot(kind='kde', ax=ax, color = "red")
ax.set_xlabel('Mean Age')
ax.set_ylabel('Density')
ax.set_title('Kernel Density Estimate of Mean Age')
plt.show()
```

#  Median Imputaion

```python
X_train["Median_Age"]  = X_train["Age"].fillna(X_train["Age"].median())
```

```python
X_train
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
X_train["Age"].plot(kind='kde', ax=ax)
X_train["Median_Age"].plot(kind='kde', ax=ax, color = "red")
ax.set_xlabel('Median Age')
# ax.set_ylabel('Density')
ax.set_title('Kernel Density Estimate of Median Age')
plt.show()
```
