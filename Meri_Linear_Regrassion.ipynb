```python
class MeriLR:
    def __init__(self):
        self.m = None
        self.b = None
    
    def fit(self, X_train, y_train):
        numo = 0
        deno = 0
        for i in range(X_train.shape[0]):
            numo = numo + (X_train[i] - X_train.mean()) * (y_train[i] - y_train.mean())
#             numo = numo + ((X_train[i] - X_train.mean()) * (y_train[i] - y_train.mean()))
    
            deno = deno +  ((X_train[i] - X_train.mean()) * (X_train[i] - X_train.mean()))
#             deno = deno + ((X_train[i] - X_train.mean())*(X_train[i] - X_train.mean()))
        
        self.m = numo/deno
        
        self.b = y_train.mean() - (self.m *  X_train.mean())
    
        print(self.m)
        print(self.b)
        
        
    def predict(self, X_test):
        
        print( self.m * X_test + self.b)
    
```
