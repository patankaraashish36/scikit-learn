```python
from sklearn.preprocessing import OrdinalEncoder , OneHotEncoder
df = pd.read_excel("C:/Users/patan/DS/Machine Learning/Dataset/global_superstore.xlsx")
```

```python
df.sample(5)
```

```python
data = df[["Segment","Category","Sub-Category",]]
data.head()
```

```python
ohe = OneHotEncoder()
fit_data = ohe.fit_transform(data).toarray()
fit_data
```

```python
features = ohe.categories_
```

```python
features = [names for sub_features in features for names in sub_features]
print(features)
```

```python
new_df = pd.DataFrame(fit_data, columns = features)
new_df.head()
```

```python
ohe_df = pd.concat([df.drop(["Segment","Category","Sub-Category",],axis = 1), new_df], axis = 1)
```
